```python
import pandas as pd
import numpy as np
```

```python
#created a csv file with state/province, country/region, lat,long & imported it to python.

file="C:/Users/Asus/Desktop/Location.csv"
```

```python
#each row in the Location.csv repeated for 60 times & output it as New_Location.csv.

output_file="C:/Users/Asus/Desktop/New_location.csv"

df=pd.read_csv(file).values
val=[]
for i in df:
    for x in range(60):
        val.append(i)
df2 = pd.DataFrame(val)
df2.to_csv(output_file,index=False)
```
